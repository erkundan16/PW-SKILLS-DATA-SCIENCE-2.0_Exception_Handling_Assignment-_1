Q1. What is an Exception in Python? Write the difference between Exceptions and Syntax errors.

An exception is an event that occurs during the execution of a program that disrupts the normal flow of instructions. In Python, exceptions are objects that represent errors. They can be generated by the interpreter or by the programmer.

Syntax errors occur when the Python parser detects an incorrect statement. These errors are typically caused by typos or mistakes in the structure of the program. Exceptions, on the other hand, occur during the execution of a program and are not typically detected by the parser.

Q2. What happens when an exception is not handled? Explain with an example.

```python
#When an exception is not handled, it results in the termination of the program and an error message is displayed. 

#For example:

import logging

logging.basicConfig(level=logging.DEBUG)

try:
    a = 5
    b = 0
    c = a / b  # This will cause a ZeroDivisionError since you cannot divide by zero
    print(c)  # This line will not be executed
except Exception as e:
    logging.error("An exception occurred: {}".format(e))

```

Q3. Which Python statements are used to catch and handle exceptions? Explain with an example.

The try, except, else, finally statements are used to catch and handle exceptions in Python.

Q4. Explain with an example:

```python
#  a. try and else:

import logging

logging.basicConfig(level=logging.DEBUG)

try:
    age = int(input("Please enter your age: "))
except ValueError as ve:
    logging.error("Invalid input. Please enter a valid integer. Error: {}".format(ve))
else:
    logging.info("I see that you are {} years old.".format(age))
```

Q5. What are Custom Exceptions in Python? Why do we need Custom Exceptions? Explain with an example.

Custom exceptions are user-defined exceptions that allow you to create specialized error classes. They can be helpful in providing more context-specific information about errors. We need custom exceptions to differentiate between different types of errors more easily and to handle them more effectively.

Q6. Create a custom exception class. Use this class to handle an exception.

```python
import logging

logging.basicConfig(level=logging.DEBUG)


class MyCustomError(Exception):
    def __init__(self, message="This is a custom error"):
        self.message = message
        super().__init__(self.message)


# Using the custom exception class
try:
    raise MyCustomError("Custom error message")
except MyCustomError as e:
    logging.error("Caught the custom error: {}".format(e))
```
